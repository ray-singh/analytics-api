```python
import asyncio
import yfinance as yf

# define your message callback
def message_handler(message):
    print(message)

async def main():
    # =======================
    # With Context Manager
    # =======================
    async with yf.AsyncWebSocket() as ws:
        await ws.subscribe(["AAPL"])
        await ws.listen(message_handler)
import nest_asyncio
nest_asyncio.apply()
asyncio.run(main())
```

```python
import time
from twelvedata import TDClient
import os
import dotenv

dotenv.load_dotenv()
TWELVEDATA_API_KEY = os.getenv("TWELVEDATA_API_KEY")

messages_history = []


def on_event(e):
    # do whatever is needed with data
    print(e)


td = TDClient(apikey=TWELVEDATA_API_KEY)
ws = td.websocket(symbols="AAPL", on_event=on_event)
ws.subscribe(['AAPL'])
ws.connect()
while True:
    ws.heartbeat()
    time.sleep(10)
```
